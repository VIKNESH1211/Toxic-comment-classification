```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
```

```python
from wordcloud import WordCloud
from wordcloud import STOPWORDS
```

```python
df = pd.read_csv("D:/Toxic Comments/train.csv")
```

```python
text =  df.comment_text.values
cloud = WordCloud(stopwords=STOPWORDS , background_color='black' , width = 3000 , height = 1800).generate(''.join(text))
plt.figure(figsize=(10,8))
plt.imshow(cloud)
```

Lowercasing

```python
df_text = df[["comment_text"]]
```

```python
df_text.head()
```

```python
df_text["comment_text"] = df_text["comment_text"].str.lower()
```

```python
df_text.head()
```

Stop Words Removal

```python
from nltk.corpus import stopwords
```

```python
stop  = stopwords.words('english')
df_text['comment_text'] = df_text['comment_text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
df_text.head()
```

```python
df["comment_text"] = df_text["comment_text"]
```

```python
df.to_csv("to_train.csv")
```
